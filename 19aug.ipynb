```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# Initialize a YOLO model
model = YOLO('yolov8s-world.pt')  # Load the YOLO-World model

# Perform prediction on the image
results = model.predict('image2.jpg', device='cpu')

# Define your custom vocabulary
custom_vocab = ["milk", "apple", "orange", "eggs"]

# Create a blank image to draw the filtered results
image = cv2.imread('image2.jpg')

# Filter the detected objects based on the custom vocabulary
for box in results[0].boxes:
    label = box.cls  # Get the class index
    class_name = results[0].names[int(label)]  # Convert index to class name
    if class_name in custom_vocab:  # Check if class name is in your custom vocabulary
        # Extract coordinates for the bounding box
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box coordinates
        confidence = box.conf[0]  # Confidence score

        # Draw the bounding box and label on the image
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Draw rectangle
        label_text = f"{class_name} ({confidence:.2f})"
        cv2.putText(image, label_text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)  # Add label

# Convert the BGR image to RGB for Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image using Matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Hide axis
plt.show()
```

pytorch

```python
from ultralytics import YOLO
import torch
import matplotlib.pyplot as plt
import cv2

# Initialize the YOLO model using the pretrained YOLOv8s-world model
model = YOLO('yolov8s-world.pt')

# Define your custom classes (these are the classes you want to detect)
custom_classes = ["milk", "apple", "orange", "egg"]

# Perform inference on an image using PyTorch
image_path = 'image1.jpg'
results = model.predict(image_path, device='cpu')

# Mapping of existing class indices to your custom classes
label_mapping = {
    "bottle": "milk",   # Assuming "bottle" represents "milk"
    "banana": "orange", # Assuming "banana" represents "orange"
    "apple": "apple",   # Direct match
    "carton": "egg"     # Assuming "carton" represents "egg"
}

# Load the image for displaying results
image = cv2.imread(image_path)

# Filter results based on the custom classes
for result in results:
    for box in result.boxes:
        label_idx = int(box.cls)  # Get the class index
        original_class_name = result.names[label_idx]  # Get the detected class name
        if original_class_name in label_mapping:
            mapped_class_name = label_mapping[original_class_name]  # Map to your custom vocabulary
            if mapped_class_name in custom_classes:
                # Extract coordinates for the bounding box
                x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box coordinates
                confidence = box.conf[0]  # Confidence score

                # Draw the bounding box and label on the image
                cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Draw rectangle
                label_text = f"{mapped_class_name} ({confidence:.2f})"
                cv2.putText(image, label_text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)  # Add label

# Save the filtered and annotated image
output_path = 'output_image.png'
cv2.imwrite(output_path, image)

# Display the image using Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')  # Hide axis
plt.show()

print(f"Filtered image saved as '{output_path}'")
```
